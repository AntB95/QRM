```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize
```

```python
correl = lambda sig: (np.exp(sig) - 1)/np.sqrt((np.exp(1) - 1) * (np.exp(sig **2) - 1))
x_pts = np.linspace(0,5,1000)

plt.figure(figsize = (16,8))
plt.plot(x_pts , [correl(x) for x in x_pts])
```

```python
scipy.optimize.fmin(-correl(x))
```
